```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
from datetime import datetime

# 1. First ensure proper datetime conversion
def safe_date_parse(x):
    try:
        return pd.to_datetime(x, format='%Y-%m-%d %H:%M:%S')  # Try common format first
    except:
        try:
            return pd.to_datetime(x)  # Fallback to automatic parsing
        except:
            return pd.NaT  # Mark unparseable dates as missing

df['Create At'] = df['Create At'].apply(safe_date_parse)

# 2. Filter out invalid dates and extract month-year
df = df[df['Create At'].notna()]
df['Month'] = df['Create At'].dt.to_period('M').astype(str)

# 3. Create widgets with sorted months
all_months = sorted(df['Month'].unique())
month_dropdown = widgets.Dropdown(
    options=['All Months'] + all_months,
    value='All Months',
    description='Filter Month:',
    style={'description_width': '150px'},
    layout={'width': '300px'}
)

output = widgets.Output()

# 4. Enhanced plotting function with error handling
def update_plot(month_filter):
    with output:
        clear_output(wait=True)
        
        try:
            # Filter data
            if month_filter == 'All Months':
                df_filtered = df.copy()
                month_title = ""
            else:
                df_filtered = df[df['Month'] == month_filter]
                month_title = f" (Month: {month_filter})"
            
            # Get top insurers for filtered data
            top_insurers = df_filtered['Insurer'].value_counts().head(8).index
            df_top = df_filtered[df_filtered['Insurer'].isin(top_insurers)]
            
            if len(df_top) == 0:
                print(f"No data available for {month_filter}")
                return
            
            # Group and calculate metrics
            grouped = df_top.groupby(['Insurer', 'Insurer Branch Name', 'Rejected']).size().unstack().reset_index()
            grouped['Total'] = grouped[False].fillna(0) + grouped[True].fillna(0)
            grouped['Approved_Pct'] = (grouped[False]/grouped['Total']*100).round(1)
            grouped['Rejected_Pct'] = (grouped[True]/grouped['Total']*100).round(1)
            grouped['Insurer_With_Branch'] = grouped['Insurer'] + '\n(' + grouped['Insurer Branch Name'] + ')'

            # Create plot
            plt.figure(figsize=(35, 15))
            x = range(len(grouped['Insurer_With_Branch']))
            width = 0.35

            # Plot bars with error handling
            plt.bar(x, grouped[False].fillna(0), width, color='skyblue', label='Approved')
            plt.bar([p + width for p in x], grouped[True].fillna(0), width, color='salmon', label='Rejected')

            # Customize plot
            plt.title(f'Top Insurers by Request Count with Approval Rates{month_title}', fontsize=24)
            plt.xlabel('Insurer (Branch)', fontsize=20)
            plt.ylabel('Number of Requests', fontsize=20)
            plt.xticks([p + width/2 for p in x], grouped['Insurer_With_Branch'], rotation=45, ha='right', fontsize=14)

            # Add value labels
            for i, row in grouped.iterrows():
                if row[False] > 0:
                    plt.text(i, row[False] + 5, f"{row[False]:.0f}\n({row['Approved_Pct']}%)", 
                            ha='center', fontsize=16, va='bottom')
                if row[True] > 0:
                    plt.text(i + width, row[True] + 5, f"{row[True]:.0f}\n({row['Rejected_Pct']}%)", 
                            ha='center', fontsize=16, va='bottom')

            plt.legend(title='Rejection Status',
                     labels=['Approved (False)', 'Rejected (True)'], 
                     fontsize=18,
                     title_fontsize=20)
            plt.grid(axis='y', linestyle='--', alpha=0.7)
            plt.tight_layout()
            plt.show()
            
        except Exception as e:
            print(f"Error generating plot: {str(e)}")

# Connect widget to function
def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        update_plot(change['new'])

month_dropdown.observe(on_change)

# Display widgets and initial plot
display(widgets.VBox([month_dropdown, output]))
update_plot('All Months')
```

```python
import pandas as pd
import plotly.graph_objects as go
from IPython.display import display, clear_output
from ipywidgets import Dropdown, HBox, VBox, Output, Button
import io

# Convert datetime
df['Create At'] = pd.to_datetime(df['Create At'], errors='coerce')
df['Month'] = df['Create At'].dt.to_period('M').astype(str)

# Setup dropdowns
manufacturer_options = ['All'] + sorted(df['Vehicle Manufacturer'].dropna().unique())
body_type_options = ['All'] + sorted(df['Vehicle Body Type'].dropna().unique())
reason_options = ['All'] + sorted(df['Rejected Reason'].dropna().unique())
insurer_options = ['All'] + sorted(df['Insurer'].dropna().unique())
branch_options = ['All'] + sorted(df['Insurer Branch Name'].dropna().unique())

manufacturer_dropdown = Dropdown(options=manufacturer_options, description='Manufacturer:')
body_type_dropdown = Dropdown(options=body_type_options, description='Body Type:')
reason_dropdown = Dropdown(options=reason_options, description='Rejected Reason:')
insurer_dropdown = Dropdown(options=insurer_options, description='Insurer:')
branch_dropdown = Dropdown(options=branch_options, description='Branch:')

output_plot = Output()
output_table = Output()

def get_filtered_df():
    temp = df.copy()

    # For each filter, create a boolean mask or all True if 'All' selected
    m_mask = (temp['Vehicle Manufacturer'] == manufacturer_dropdown.value) if manufacturer_dropdown.value != 'All' else pd.Series([True]*len(temp), index=temp.index)
    b_mask = (temp['Vehicle Body Type'] == body_type_dropdown.value) if body_type_dropdown.value != 'All' else pd.Series([True]*len(temp), index=temp.index)
    r_mask = (temp['Rejected Reason'] == reason_dropdown.value) if reason_dropdown.value != 'All' else pd.Series([True]*len(temp), index=temp.index)
    i_mask = (temp['Insurer'] == insurer_dropdown.value) if insurer_dropdown.value != 'All' else pd.Series([True]*len(temp), index=temp.index)
    br_mask = (temp['Insurer Branch Name'] == branch_dropdown.value) if branch_dropdown.value != 'All' else pd.Series([True]*len(temp), index=temp.index)

    # Rows matching ANY filter (OR across filters)
    combined_mask = m_mask | b_mask | r_mask | i_mask | br_mask

    filtered = temp[combined_mask].copy()

    # Blank out values where mask is False (aligned with filtered's index)
    if manufacturer_dropdown.value != 'All':
        filtered.loc[~m_mask.loc[filtered.index], 'Vehicle Manufacturer'] = ''
    if body_type_dropdown.value != 'All':
        filtered.loc[~b_mask.loc[filtered.index], 'Vehicle Body Type'] = ''
    if reason_dropdown.value != 'All':
        filtered.loc[~r_mask.loc[filtered.index], 'Rejected Reason'] = ''
    if insurer_dropdown.value != 'All':
        filtered.loc[~i_mask.loc[filtered.index], 'Insurer'] = ''
    if branch_dropdown.value != 'All':
        filtered.loc[~br_mask.loc[filtered.index], 'Insurer Branch Name'] = ''

    # Drop duplicates by 'Variant ID', keep the first occurrence
    if 'Variant ID' in filtered.columns:
        filtered = filtered.drop_duplicates(subset=['Variant ID'])

    return filtered

def update_output(*args):
    filtered_df = get_filtered_df()

    with output_plot:
        clear_output(wait=True)

        # Group and count on filtered_df for plotting
        agg_data = filtered_df.groupby(
            ['Rejected Reason', 'Insurer', 'Insurer Branch Name']
        ).size().reset_index(name='Count')

        if agg_data.empty:
            print("No data available for selected filters.")
            return

        # Top 20 combinations
        top_combos = agg_data.sort_values('Count', ascending=False).head(20)
        top_combos['Label'] = top_combos['Rejected Reason'] + '<br>' + top_combos['Insurer'] + ' - ' + top_combos['Insurer Branch Name']

        # Plot
        fig = go.Figure(go.Bar(
            x=top_combos['Count'],
            y=top_combos['Label'],
            orientation='h',
            marker_color='#1f77b4',
            hovertemplate='<b>%{y}</b><br>Count: %{x}<extra></extra>'
        ))
        fig.update_layout(
            title='Top 20 Rejection Reasons by Count',
            height=800,
            yaxis=dict(autorange="reversed", tickfont=dict(size=12)),
            margin=dict(l=200, r=50, t=80, b=50)
        )
        fig.show()

    with output_table:
        clear_output(wait=True)

        # Calculate frequency for each row's combo (for display)
        freq_series = filtered_df.groupby(
            ['Rejected Reason', 'Insurer', 'Insurer Branch Name']
        ).size()

        # Map frequency count to each row
        filtered_df['Frequency'] = filtered_df.apply(
            lambda row: freq_series.get((row['Rejected Reason'], row['Insurer'], row['Insurer Branch Name']), 0),
            axis=1
        )

        columns_to_show = [
            'Vehicle Manufacturer', 'Model Name', 'Vehicle Variant', 'Make', 'Model', 'Variant',
            'Make Code', 'Model Code', 'Variant Code', 'Variant ID', 'New Vehicle', 'Fuel Type',
            'Min IDV', 'Max IDV', 'Vehicle IDV', 'Frequency'
        ]
        existing_columns = [col for col in columns_to_show if col in filtered_df.columns]

        if not filtered_df.empty and existing_columns:
            display(
                filtered_df[existing_columns].head(1000).style.set_caption("Detailed View with Frequency (First 100 Rows)")
            )
        else:
            print("No matching detailed data found.")

# Export button to CSV
def export_table(_):
    filtered_df = get_filtered_df()
    filtered_df.to_csv('filtered_output.csv', index=False)
    print("Exported filtered data to 'filtered_output.csv'.")

export_button = Button(description='Export Table as CSV')
export_button.on_click(export_table)

# Link dropdowns to update function
for dropdown in [manufacturer_dropdown, body_type_dropdown, reason_dropdown, insurer_dropdown, branch_dropdown]:
    dropdown.observe(update_output, names='value')

# Show UI
display(HBox([manufacturer_dropdown, body_type_dropdown, reason_dropdown]))
display(HBox([insurer_dropdown, branch_dropdown]))
display(export_button)
display(output_plot)
display(output_table)

# Initial render
update_output()
```

```python
#-----------------------------------------------------------FINAL CODE FOR QOUTELOG ERROR ANALYSIS---------------------------------------------------------------------------------------
import pandas as pd
import plotly.graph_objects as go
from IPython.display import display, clear_output
from ipywidgets import Dropdown, HBox, VBox, Output, Button
from io import StringIO

# --- Load and prepare the data ---
# Replace with your actual data source
# df = pd.read_csv('your_file.csv')  # Example: load data
# df = pd.DataFrame(...)             # Or create dummy data

# --- Clean and preprocess ---
df.columns = df.columns.str.strip()
df['Create At'] = pd.to_datetime(df['Create At'], errors='coerce')
df['Month'] = df['Create At'].dt.to_period('M').astype(str)

# --- Dropdown options ---
manufacturer_options = ['All'] + sorted(df['Vehicle Manufacturer'].dropna().unique())
body_type_options = ['All'] + sorted(df['Vehicle Body Type'].dropna().unique())
reason_options = ['All'] + sorted(df['Rejected Reason'].dropna().unique())
insurer_options = ['All'] + sorted(df['Insurer'].dropna().unique())
branch_options = ['All'] + sorted(df['Insurer Branch Name'].dropna().unique())
rejected_options = ['All'] + sorted(df['Rejected'].dropna().unique()) if 'Rejected' in df.columns else []

# --- Widgets ---
manufacturer_dropdown = Dropdown(options=manufacturer_options, description='Manufacturer:')
body_type_dropdown = Dropdown(options=body_type_options, description='Body Type:')
reason_dropdown = Dropdown(options=reason_options, description='Rejected Reason:')
insurer_dropdown = Dropdown(options=insurer_options, description='Insurer:')
branch_dropdown = Dropdown(options=branch_options, description='Branch:')
rejected_dropdown = Dropdown(options=rejected_options, description='Rejected:') if rejected_options else None

export_button = Button(description='Export Table to CSV', button_style='success')

output_plot = Output()
output_table = Output()
export_output = Output()

current_table_data = pd.DataFrame()

# --- Helper: Filtered Data ---
def get_filtered_df():
    temp = df.copy()
    if manufacturer_dropdown.value != 'All':
        temp = temp[temp['Vehicle Manufacturer'] == manufacturer_dropdown.value]
    if body_type_dropdown.value != 'All':
        temp = temp[temp['Vehicle Body Type'] == body_type_dropdown.value]
    if reason_dropdown.value != 'All':
        temp = temp[temp['Rejected Reason'] == reason_dropdown.value]
    if insurer_dropdown.value != 'All':
        temp = temp[temp['Insurer'] == insurer_dropdown.value]
    if branch_dropdown.value != 'All':
        temp = temp[temp['Insurer Branch Name'] == branch_dropdown.value]
    if rejected_dropdown and rejected_dropdown.value != 'All':
        temp = temp[temp['Rejected'] == rejected_dropdown.value]

    return temp

# --- Update Chart + Table ---
def update_output(*args):
    global current_table_data
    filtered_df = get_filtered_df()

    with output_plot:
        clear_output(wait=True)
        agg_data = filtered_df.groupby(
            ['Rejected Reason', 'Insurer', 'Insurer Branch Name']
        ).size().reset_index(name='Count')

        if agg_data.empty:
            print("No data available for selected filters.")
        else:
            top_combos = agg_data.sort_values('Count', ascending=False).head(50)
            top_combos['Label'] = top_combos['Rejected Reason'] + '<br>' + top_combos['Insurer'] + ' - ' + top_combos['Insurer Branch Name']

            fig = go.Figure(go.Bar(
                x=top_combos['Count'],
                y=top_combos['Label'],
                orientation='h',
                marker_color='#1f77b4',
                hovertemplate='<b>%{y}</b><br>Count: %{x}<extra></extra>'
            ))
            fig.update_layout(
                title='Top 20 Rejection Reasons by Count',
                height=800,
                yaxis=dict(autorange="reversed", tickfont=dict(size=12)),
                margin=dict(l=200, r=50, t=80, b=50)
            )
            fig.show()

    with output_table:
        clear_output(wait=True)
        current_table_data = pd.DataFrame()  # Reset before processing
        
        if filtered_df.empty:
            print("No matching data found for selected filters.")
            return

        # Get top combinations if available
        if not agg_data.empty:
            mask = pd.merge(
                filtered_df,
                agg_data[['Rejected Reason', 'Insurer', 'Insurer Branch Name']],
                on=['Rejected Reason', 'Insurer', 'Insurer Branch Name'],
                how='inner'
            )
        else:
            mask = filtered_df.copy()

        columns_to_show = [
            'Vehicle Manufacturer', 'Model Name', 'Vehicle Variant', 'Make', 'Model', 'Variant',
            'Make Code', 'Model Code', 'Variant Code', 'Variant ID', 'New Vehicle', 'Fuel Type',
            'Min IDV', 'Max IDV', 'Vehicle IDV'
        ]
        existing_columns = [col for col in columns_to_show if col in mask.columns]
        
        if not existing_columns:
            print("No matching columns found in data.")
            return

        # Create table DataFrame safely
        if not mask.empty:
            table_df = mask[existing_columns].copy()
            try:
                # Only calculate frequency if we have data
                if not table_df.empty:
                    table_df['Frequency'] = table_df.groupby(existing_columns)[existing_columns[0]].transform('size')
                    table_df = table_df.drop_duplicates().sort_values('Frequency', ascending=False)
            except Exception as e:
                print(f"Error calculating frequencies: {str(e)}")
                table_df = mask[existing_columns].copy()  # Fallback to raw data
        else:
            table_df = pd.DataFrame(columns=existing_columns)

        current_table_data = table_df.copy()

        if not table_df.empty:
            display(table_df.head(100000))
        else:
            print("No matching detailed data found.")

# --- Export to CSV ---
def export_table_to_csv(btn):
    with export_output:
        clear_output()
        if not current_table_data.empty:
            # Define filename and location
            filename = "rejection_analysis_export.csv"  # Default filename
            filepath = "./"  # Current directory
            
            # You could make these dynamic based on filters:
            if manufacturer_dropdown.value != 'All':
                filename = f"rejection_analysis_{manufacturer_dropdown.value}.csv"
            
            # Save to file
            full_path = filepath + filename
            current_table_data.to_csv(full_path, index=False)
            print(f"File saved to: {full_path}")
            
            # Also keep the downloadable version
            csv_buffer = StringIO()
            current_table_data.to_csv(csv_buffer, index=False)
            display({
                "text/csv": csv_buffer.getvalue()
            }, raw=True)
        else:
            print("No data to export.")

# --- Wire up interactions ---
widgets = [
    manufacturer_dropdown, body_type_dropdown, reason_dropdown,
    insurer_dropdown, branch_dropdown
]
if rejected_dropdown:
    widgets.append(rejected_dropdown)

for dropdown in widgets:
    dropdown.observe(update_output, names='value')

export_button.on_click(export_table_to_csv)

# --- Render UI ---
display(HBox([manufacturer_dropdown, body_type_dropdown]))
display(HBox([reason_dropdown, insurer_dropdown, branch_dropdown]))
if rejected_dropdown:
    display(HBox([rejected_dropdown]))
display(output_plot)
display(output_table)
display(export_button)
display(export_output)

# --- Initial update ---
update_output()
```
